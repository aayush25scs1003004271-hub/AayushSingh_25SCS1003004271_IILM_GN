```
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics import classification_report
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.pipeline import Pipeline

# -----------------------------
# Load your smartwatch dataset
# -----------------------------
data = pd.read_csv("smartwatch.csv")   # change your file name here

# View data
print(data.head())

# ----------------------------------------
# Preprocess price column (remove ₹ , , )
# ----------------------------------------
data["price"] = (
    data["price"].str.replace("₹", "", regex=False)
                  .str.replace(",", "", regex=False)
                  .astype(float)
)

# ----------------------------------------
# Create rating labels for classification
# Example: high rating (>=60) vs low rating (<60)
# ----------------------------------------
data["rating_label"] = np.where(data["rating"] >= 60, "High", "Low")

# Drop rows with missing labels
data = data.dropna(subset=["rating_label", "features"])

# ----------------------------------------
# Feature + Target split
# ----------------------------------------
X_text = data["features"]          # text feature
X_price = data["price"]             # numeric feature
y = data["rating_label"]

# Combine text + numeric features using pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler

preprocess = ColumnTransformer(
    transformers=[
        ("text", TfidfVectorizer(), "features"),
        ("num", StandardScaler(), ["price"])
    ]
)

# -----------------------------
# Choose a classification model
# -----------------------------
model = SVC()                      # or KNeighborsClassifier(), RandomForestClassifier()

clf = Pipeline([
    ("preprocess", preprocess),
    ("model", model)
])

# -----------------------------
# Train-test split
# -----------------------------
X_train, X_test, y_train, y_test = train_test_split(
    data[["features", "price"]],
    y,
    test_size=0.2,
    random_state=42
)

# -----------------------------
# Train the model
# -----------------------------
clf.fit(X_train, y_train)

# -----------------------------
# Predictions
# -----------------------------
pred = clf.predict(X_test)

# -----------------------------
# Evaluation
# -----------------------------
print("\nClassification Report:")
print(classification_report(y_test, pred))
```

```
# ================================================
#   SMARTWATCH CLASSIFICATION: SVM vs KNN
# ================================================
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt

# -----------------------------------------------
# Load your dataset
# -----------------------------------------------
# Replace with your CSV file name:
# data = pd.read_csv("smartwatch.csv")

# SAMPLE dataset based on your given data:
data = pd.DataFrame({
    "name": [
        "Noise NoiseFit Endeavour Pro Smartwatch",
        "Fire Boltt Oracle WristPhone",
        "Samsung Galaxy Watch 7 LTE 44mm",
        "Noise NoiseFit Endeavour 2 Smartwatch",
        "Realme Watch S2"
    ],
    "price": ["₹9,999", "₹2,999", "₹16,496", "₹5,999", "₹2,699"],
    "rating": [62.0, np.nan, np.nan, 62.0, 57.0],
    "features": [
        "Bluetooth, GPS, Bluetooth Calling",
        "Wi-Fi, Bluetooth 5.0, GPS, 3G, 4G, Voice Calling",
        "Wi-Fi, Bluetooth 5.3, GPS, NFC, 3G, 4G, VoLTE",
        "Bluetooth, GPS, Bluetooth Calling",
        "Bluetooth 5.3, Bluetooth Calling"
    ]
})

# -----------------------------------------------
# Clean price column
# -----------------------------------------------
data["price"] = (
    data["price"].str.replace("₹", "", regex=False)
                  .str.replace(",", "", regex=False)
                  .astype(float)
)

# -----------------------------------------------
# Create classification label (High / Low)
# -----------------------------------------------
data["rating_label"] = np.where(data["rating"] >= 60, "High", "Low")

# Drop rows with missing target
data = data.dropna(subset=["rating_label"])

# Features and labels
X = data[["features", "price"]]
y = data["rating_label"]

# -----------------------------------------------
# Preprocessing pipeline (TF-IDF + Scaling)
# -----------------------------------------------
preprocess = ColumnTransformer(
    transformers=[
        ("text", TfidfVectorizer(), "features"),
        ("num", StandardScaler(), ["price"])
    ]
)

# -----------------------------------------------
# Models: SVM and KNN
# KNN uses k=1 since dataset is very small
# -----------------------------------------------
models = {
    "SVM": SVC(),
    "KNN": KNeighborsClassifier(n_neighbors=1)
}

accuracies = {}

# -----------------------------------------------
# Train + Evaluate both models
# -----------------------------------------------
for model_name, model in models.items():
    pipeline = Pipeline([
        ("preprocess", preprocess),
        ("model", model)
    ])

    X_train, X_test, y_train, y_test = train_test_split(
        X, y, test_size=0.4, random_state=42
    )

    pipeline.fit(X_train, y_train)
    preds = pipeline.predict(X_test)

    accuracies[model_name] = accuracy_score(y_test, preds)

# -----------------------------------------------
# Plot comparison graph
# -----------------------------------------------
plt.figure(figsize=(6, 4))
plt.bar(accuracies.keys(), accuracies.values())
plt.title("SVM vs KNN Accuracy Comparison")
plt.ylabel("Accuracy")
plt.ylim(0, 1)
plt.grid(axis="y", linestyle="--")
plt.show()
```
